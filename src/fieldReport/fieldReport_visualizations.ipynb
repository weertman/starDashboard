```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import date2num
from scipy import stats
import os

def add_day_night_regression(ax, date_time, n_stars, night_day):
    day_night_info = {
        'day': {'color': 'orange', 'label': 'Day'},
        'night': {'color': 'blue', 'label': 'Night'}
    }
    
    regression_results = {}
    
    for period in ['day', 'night']:
        mask = night_day == period
        x = date2num(date_time[mask])
        y = n_stars[mask]
        
        if len(x) > 2:
            try:
                slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
                regression_results[period] = {
                    'slope': slope,
                    'intercept': intercept,
                    'r_squared': r_value**2,
                    'p_value': p_value,
                    'std_err': std_err,
                    'n': len(x)
                }
                
                x_range = np.linspace(date2num(date_time.min()), date2num(date_time.max()), 100)
                y_range = slope * x_range + intercept
                
                label = f"{day_night_info[period]['label']} (R = {r_value**2:.3f})"
                ax.plot(x_range, y_range, color=day_night_info[period]['color'], 
                        linestyle='--', label=label)
            except Exception as e:
                print(f"Error in regression calculation for {period}: {str(e)}")
        else:
            print(f"Not enough data points for {period} regression")
    
    if regression_results:
        ax.legend(frameon=False, fontsize=10, loc='upper right')
    
    return regression_results

def compare_slopes(day_results, night_results):
    slope_diff = day_results['slope'] - night_results['slope']
    se_diff = np.sqrt(day_results['std_err']**2 + night_results['std_err']**2)
    
    df = day_results['n'] + night_results['n'] - 4  # Corrected degrees of freedom
    t_stat = slope_diff / se_diff
    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))
    
    # Calculate confidence interval for slope difference
    ci_95 = se_diff * stats.t.ppf(0.975, df)
    
    return {
        'slope_difference': slope_diff,
        't_statistic': t_stat,
        'p_value': p_value,
        'degrees_of_freedom': df,
        'confidence_interval_95': (slope_diff - ci_95, slope_diff + ci_95)
    }

# Main code
target_survey = 'site1__circle'
tmp = survey_record[survey_record['survey'] == target_survey]
n_stars = tmp['num_stars']
night_day_colors = [night_day_color[i] for i in tmp['night_day']]

ylims = [-.2, max(n_stars)+1]
yticks = sorted(list(set(n_stars)))

xticks = tmp['date_time']
xlabs = [i.strftime('%m-%d-%Y %H:%M') for i in xticks]

fig, ax = plt.subplots(1,1,figsize=(8,4))

ax.scatter(tmp['date_time'], n_stars, color=night_day_colors, s=50, edgecolor='k', alpha=0.7)

regression_results = add_day_night_regression(ax, tmp['date_time'], n_stars, tmp['night_day'])

ax.set_xticks(xticks)
ax.set_xticklabels(xlabs, rotation=45, ha='right')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_ylabel('# Stars Observed')
ax.set_ylim(ylims)
ax.set_yticks(yticks)

ax.set_title(f'Circle survey\n2-year old sunflower star observations at site 1', loc='left')

path_fig = os.path.join(target_dir, f'{target_survey}_sunflower_star_observation.png')
plt.savefig(path_fig, dpi=300, bbox_inches='tight')

plt.tight_layout()
plt.show()
plt.close()

# Print regression statistics
for period, results in regression_results.items():
    print(f"\n{period.capitalize()} regression statistics:")
    for stat, value in results.items():
        print(f"  {stat}: {value}")

# Compare slopes
slope_comparison = compare_slopes(regression_results['day'], regression_results['night'])
print("\nSlope Comparison Results:")
for key, value in slope_comparison.items():
    print(f"  {key}: {value}")

# Perform Mann-Whitney U test
day_data = n_stars[tmp['night_day'] == 'day']
night_data = n_stars[tmp['night_day'] == 'night']
statistic, p_value = stats.mannwhitneyu(day_data, night_data, alternative='two-sided')

print("\nMann-Whitney U test results:")
print(f"  Statistic: {statistic}")
print(f"  p-value: {p_value}")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Assuming the dataframe is already loaded as df
# If not, uncomment the following line:
# df = pd.read_csv('your_file.csv')

# Create a contingency table
contingency_table = pd.crosstab(df['is_daytime'], df['Activity'])

# Perform chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print results
print("Contingency Table:")
print(contingency_table)
print("\nChi-square statistic:", chi2)
print("p-value:", p_value)
print("Degrees of freedom:", dof)

# Interpret the results
alpha = 0.05  # Common significance level
if p_value < alpha:
    print("\nReject the null hypothesis.")
    print("There is a significant association between time of day and activity.")
else:
    print("\nFail to reject the null hypothesis.")
    print("There is no significant evidence of an association between time of day and activity.")

# Calculate Cramer's V for effect size
n = contingency_table.sum().sum()
min_dim = min(contingency_table.shape) - 1
cramer_v = np.sqrt(chi2 / (n * min_dim))

print("\nCramer's V:", cramer_v)
print("Effect size interpretation:")
if cramer_v < 0.1:
    print("Negligible association")
elif cramer_v < 0.3:
    print("Weak association")
elif cramer_v < 0.5:
    print("Moderate association")
else:
    print("Strong association")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import fisher_exact

# Assuming the dataframe is already loaded as df
# If not, uncomment the following line:
# df = pd.read_csv('your_file.csv')

# Create a contingency table
contingency_table = pd.crosstab(df['is_daytime'], df['Activity'])

# Perform Fisher's Exact Test
odds_ratio, p_value = fisher_exact(contingency_table)

# Print results
print("Contingency Table:")
print(contingency_table)
print("\nFisher's Exact Test Results:")
print("Odds Ratio:", odds_ratio)
print("p-value:", p_value)

# Interpret the results
alpha = 0.05  # Common significance level
if p_value < alpha:
    print("\nReject the null hypothesis.")
    print("There is a significant association between time of day and activity.")
else:
    print("\nFail to reject the null hypothesis.")
    print("There is no significant evidence of an association between time of day and activity.")

# Calculate Phi coefficient
a = contingency_table.iloc[0, 0]  # Night Active
b = contingency_table.iloc[0, 1]  # Night Resting
c = contingency_table.iloc[1, 0]  # Day Active
d = contingency_table.iloc[1, 1]  # Day Resting
n = a + b + c + d

phi = np.sqrt((a*d - b*c)**2 / ((a+b)*(c+d)*(a+c)*(b+d)))

print("\nPhi Coefficient:", phi)
print("Effect size interpretation:")
if phi < 0.1:
    print("Negligible association")
elif phi < 0.3:
    print("Weak association")
elif phi < 0.5:
    print("Moderate association")
else:
    print("Strong association")
```
